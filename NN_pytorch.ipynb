```python
import torch
from torch.autograd import Variable
```

```python
dtype=torch.FloatTensor

N,D_in,H,F_out = 64,1000,100,10
x=Variable(torch.randn(N,D_in).type(dtype),requires_grad=False)
y=Variable(torch.randn(N,F_out).type(dtype),requires_grad=False)
  
w1=Variable(torch.randn(D_in,H).type(dtype),requires_grad=True)
w2=Variable(torch.randn(H,F_out).type(dtype),requires_grad=True)
 

lr_rate=1e-6
for t in range(500):
    y_pred = x.mm(w1).clamp(min=0).mm(w2)
    loss= (y_pred-y).pow(2).sum()
    print(t,loss.data[0])
    loss.backward()
    w1.data=w1.data - lr_rate*(w1.grad.data)
    w2.data=w2.data -lr_rate*(w2.grad.data)
    w1.grad.data.zero_()
    w2.grad.data.zero_()

    
```
